```python
def tuple_str(t) :
    res = ""
    for x in t : res += x
    return res
```

```python
def get_combi_idx_pos(config_history) :
    combi_idx = dict(); combi_pos = dict()
    i = 0
    for gen, configs in config_history.items() :
        c = tuple_str(tuple(configs))
        if c not in combi_idx.keys() :
            combi_idx[c] = chr(65 + i)
            combi_pos[chr(65 + i)] = [gen]
            i += 1
        else : 
            combi_pos[combi_idx[c]].append(gen)
    return combi_idx, combi_pos
```
