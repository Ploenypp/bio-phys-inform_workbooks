```r
# import libraries 
library(ggplot2)
library(tidyverse)
library(ggfortify)
library(ggrepel)
library(factoextra)
library(FactoMineR)
```

```r
sharks = read.table("datasets/shark_immunity.txt", header = TRUE, sep = "\t")

# change the order the physiological conditions are displayed 
sharks$physiological_condition <- factor(
  sharks$physiological_condition,
  levels = c("Healthy", "Stressed", "Injured", "Infected", "Recovering")
)
```

```r
options(repr.plot.width=8, repr.plot.height=5)
```

```r
condition_data = function(data, condition) {
    data[data$physiological_condition==condition, ]
}

cell_data = function(data, cell_type) {
    data[data$immune_cell_type==cell_type, ]
}
```

```r
healthy = condition_data(sharks, "Healthy")
stressed = condition_data(sharks, "Stressed")
injured = condition_data(sharks, "Injured")
infected = condition_data(sharks, "Infected")
recovering = condition_data(sharks, "Recovering")

healthy_pca = PCA(healthy[ ,4:9], scale.unit=TRUE, graph=FALSE)
stressed_pca = PCA(stressed[ ,4:9], scale.unit=TRUE, graph=FALSE)
injured_pca = PCA(injured[ ,4:9], scale.unit=TRUE, graph=FALSE)
infected_pca = PCA(infected[ ,4:9], scale.unit=TRUE, graph=FALSE)
recovering_pca = PCA(recovering[ ,4:9], scale.unit=TRUE, graph=FALSE)
```

```r
fviz_pca_ind(healthy_pca, col.ind=healthy$immune_cell_type, addEllipses = TRUE, title="Healthy PCA")

fviz_pca_ind(stressed_pca, col.ind=stressed$immune_cell_type, addEllipses = TRUE, title="Stressed PCA")

fviz_pca_ind(injured_pca, col.ind=injured$immune_cell_type, addEllipses = TRUE, title="Injured PCA")

fviz_pca_ind(infected_pca, col.ind=infected$immune_cell_type, addEllipses = TRUE, title="Infected PCA")

fviz_pca_ind(recovering_pca, col.ind=recovering$immune_cell_type, addEllipses = TRUE, title="Recovering PCA")
```

```r
fviz_pca_var(healthy_pca, title="Healthy PCA")
fviz_pca_var(stressed_pca, title="Stressed PCA")
fviz_pca_var(infected_pca, title="Infected PCA")
fviz_pca_var(injured_pca, title="Injured PCA")
fviz_pca_var(recovering_pca, title="Recovering PCA")
```

```r
immune_pca = PCA(sharks[,12:14], scale.unit=TRUE, graph=FALSE)
fviz_pca_ind(immune_pca, col.ind=sharks$immune_cell_type, addEllipses = TRUE, title="Immune Response PCA")
fviz_pca_var(immune_pca, title="Immune Response PCA")
```

```r
immune_pca = PCA(recovering[,12:14], scale.unit=TRUE, graph=FALSE)
fviz_pca_ind(immune_pca, col.ind=recovering$immune_cell_type, addEllipses=TRUE, title="Immune Response PCA for Immune Cell Type?")
fviz_pca_var(immune_pca, title="Immune Response PCA for Immune Cell Type?")
```
