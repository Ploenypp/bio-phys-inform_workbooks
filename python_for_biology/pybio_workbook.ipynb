```python
def tuple_str(t) :
    res = ""
    for x in t : res += x
    return res
```

```python
def get_config_pos(config_history) :
    combi_pos = dict()
    for gen, configs in config_history.items() :
        for c in configs :
            if c not in combi_pos.keys() : combi_pos[c] = [gen]
            else : combi_pos[c].append(gen)
    return combi_pos
            
```

```python
config_pos = get_config_pos(config_history)
config_pos
```

```python
def find_arith_prog(config_pos) :
    # return the starting index, the step size, and the last index
    
    config_arith = dict()
    for config, pos in config_pos.items() :
        if len(pos) > 1 : 
            arith = np.gradient(pos)
            if all(x == arith[0] for x in arith) : config_arith[config] = (pos[0], int(arith[0]), pos[-1])
            continue

        config_arith[config] = (pos[0], 0, pos[-1])
        
    return config_arith
```
