```python
def config_history_toSeq(config_history) :
    return [x for lst in config_history.values() for x in lst]
```

```python
def get_min_max_config_gen(config_history) :
    sizes = [len(lst) for lst in config_history.values()]
    return min(sizes), max(sizes)
```

```python
def hashtable(hist_seq, length) :
    ht = dict()
    for i in range(len(hist_seq)-length) :
        substr = ""
        for j in range(i,i+length) : 
            substr += str(hist_seq[j])
            if j < i+length : substr += " "
        if substr not in ht.keys() : ht[substr] = 0
        ht[substr] += 1
    return ht
```
